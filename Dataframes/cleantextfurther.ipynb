```python
import pandas as pd
import re
import html

def clean_text(text):
    if not isinstance(text, str):
        return text

    text = html.unescape(text) # Decode HTML entities
    text = text.replace('\n', ' ').replace('\t', ' ')
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.replace('\xa0', ' ')  # Non-breaking spaces
    text = text.lower() # Convert to lowercase
    # Remove emojis (more robust method)
    text = re.sub(r'[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F1E0-\U0001F1FF]+', '', text)
    text = text.strip()
    return text

def clean_csv(input_file, output_file):
    """Cleans the 'Text' column of a CSV file and writes to a new CSV."""
    try:
        df = pd.read_csv(input_file)
    except FileNotFoundError:
        print(f"Error: Input file '{input_file}' not found.")
        return
    except pd.errors.ParserError:
        print(f"Error: Could not parse input file '{input_file}'. Check the CSV format.")
        return

    if 'Text' not in df.columns:
        print("Error: Input CSV must contain 'Text' column.")
        return

    df['Text'] = df['Text'].apply(clean_text)

    try:
        df.to_csv(output_file, index=False)
        print(f"Successfully wrote cleaned data to '{output_file}'")
    except Exception as e:
        print(f"Error writing to output file: {e}")


# Example usage:
input_csv = r"C:\Users\jatro\Dev\Dataframes\clean_conversations.csv"  # Replace with your input file name
output_csv = r"C:\Users\jatro\Dev\Dataframes\cleaner_conversations.csv"  # Replace with desired output file name
clean_csv(input_csv, output_csv)












#def clean_df(df):
#    df['Text'] = df['Text'].apply(clean_text)
#    return df
```
