```python
import pandas as pd
import re

def clean_text(text):
    """Removes unnecessary spaces, newlines, and tabs from a string."""

    if not isinstance(text, str): #check if the value is a string
        return text

    # Remove newline characters and tabs
    text = text.replace('\n', ' ').replace('\t', ' ')

    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove non-breaking spaces ( )
    text = text.replace('\xa0', ' ')

    #Remove any leading or trailing spaces after all other cleaning
    text = text.strip()

    return text

def clean_csv(input_file, output_file):
    """Cleans the 'Text' column of a CSV file and writes to a new CSV."""
    try:
        df = pd.read_csv(input_file)
    except FileNotFoundError:
        print(f"Error: Input file '{input_file}' not found.")
        return
    except pd.errors.ParserError:
        print(f"Error: Could not parse input file '{input_file}'. Check the CSV format.")
        return

    if 'Text' not in df.columns:
        print("Error: Input CSV must contain 'Text' column.")
        return

    df['Text'] = df['Text'].apply(clean_text)

    try:
        df.to_csv(output_file, index=False)
        print(f"Successfully wrote cleaned data to '{output_file}'")
    except Exception as e:
        print(f"Error writing to output file: {e}")


# Example usage:
input_csv = r"C:\Users\jatro\Dev\Dataframes\concatenated_conversations.csv"  # Replace with your input file name
output_csv = r"C:\Users\jatro\Dev\Dataframes\clean_conversations.csv"  # Replace with desired output file name
clean_csv(input_csv, output_csv)
```
