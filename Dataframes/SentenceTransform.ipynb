```python
from sentence_transformers import SentenceTransformer
import json
from tqdm import tqdm

try:
    model = SentenceTransformer('all-mpnet-base-v2')
except Exception as e:
    print(f"Error loading model: {e}")
    exit()

try:
    with open('RAG_conv.json', 'r', encoding='utf-8') as f:
        conversations = json.load(f)
except FileNotFoundError:
    print("Error: RAG_conv.json not found.")
    exit()
except json.JSONDecodeError:
    print("Error: Invalid JSON format in RAG_conv.json.")
    exit()

embedded_conversations = {}
for thread_id, messages in tqdm(conversations.items(), desc="Embedding threads"):
    embedded_conversations[thread_id] = []
    message_texts = [message_data.get('message', '') for message_data in messages] #Handles missing message key
    try:
        embeddings = model.encode(message_texts).tolist()
    except Exception as e:
        print(f"Error during embedding: {e}")
        continue #Skip to next thread

    for i, message_data in enumerate(messages):
        embedded_conversations[thread_id].append({
            "sender": message_data.get('sender', ''), #Handles missing sender key
            "timestamp": message_data.get('timestamp', ''), #Handles missing timestamp key
            "message": message_data.get('message', ''),
            "embedding": embeddings[i]
        })

try:
    with open('embedded_conversations.json', 'w', encoding='utf-8') as outfile:
        json.dump(embedded_conversations, outfile, indent=4, ensure_ascii=False)
    print("Embeddings saved to embedded_conversations.json")
except Exception as e:
    print(f"Error saving JSON: {e}")
```
