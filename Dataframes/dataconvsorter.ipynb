```python
import pandas as pd

def sort_data(input_file, output_file):
    """
    Sorts data by Thread ID and then by 'Created At' date.

    Args:
        input_file: Path to the input CSV file.
        output_file: Path to the output CSV file.
    """
    try:
        df = pd.read_csv(input_file)
    except FileNotFoundError:
        print(f"Error: Input file '{input_file}' not found.")
        return
    except pd.errors.ParserError:
        print(f"Error: Could not parse input file '{input_file}'. Check the CSV format.")
        return

    if 'Thread ID' not in df.columns or 'Created At' not in df.columns:
        print("Error: Input CSV must contain 'Thread ID' and 'Created At' columns.")
        return
    
    try:
        df['Created At'] = pd.to_datetime(df['Created At'], utc=True)
    except pd.errors.OutOfBoundsDatetime:
        print("Error: Could not parse 'Created At' column. Check date format.")
        return

    # Sort by Thread ID and then by Created At
    df_sorted = df.sort_values(by=['Thread ID', 'Created At'])

    try:
        df_sorted.to_csv(output_file, index=False)
        print(f"Successfully wrote sorted data to '{output_file}'")
    except Exception as e:
        print(f"Error writing to output file: {e}")


# Example usage:
input_csv = "input.csv"  # Replace with your input file name
output_csv = "output.csv"  # Replace with desired output file name
sort_data(r'C:\\Users\\jatro\\Dev\\Dataframes\\filtered_conversations.csv', r'C:\\Users\\jatro\\Dev\\Dataframes\\datesorted_conversations.csv')
```
